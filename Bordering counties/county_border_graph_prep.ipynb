```python
import pandas as pd
import json
import csv
```

```python
c_borders = json.load(open('countycodes_border.json', 'r'))
c_borders
```

```python
borders_csv = open('counties_borders_csv_weight','w')
graph_writer = csv.writer(borders_csv, delimiter=',', quotechar='"')

for k, v in c_borders.items():
    for V in v:
        graph_writer.writerow([k, V, 1])
borders_csv.close()
df = pd.read_csv('counties_borders_csv_weight', header=None)
df
```
