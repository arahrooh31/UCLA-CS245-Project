```python
import pandas as pd
from sodapy import Socrata

client = Socrata("data.cdc.gov", None)

# https://data.cdc.gov/Public-Health-Surveillance/United-States-COVID-19-County-Level-of-Community-T/nra9-vzzn
filename_covid_cases_state_county = '/content/gdrive/MyDrive/COVID DGNN KG/CDC_COVID19_cases_and_vaccine_time_series/covid_cases_by_county.csv'
fileid_covid_cases_state_county = "nra9-vzzn"

# converted to Python list of dictionaries by sodapy
results = client.get(fileid_covid_cases_state_county, where="date >= '2020-03-01T00:00:00.000'", limit = 3000000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)

results_df.to_csv(filename_covid_cases_state_county)

```
